```python
import gensim.downloader as api
```

```python
model = api.load('fasttext-wiki-news-subwords-300') # 1M voc, 300d
```

```python
print(model.most_similar('paris', topn=10))
# Analogies: paris - france + germany ≈ ?
vec = model['paris'] - model['france'] + model['germany']
print(model.similar_by_vector(vec, topn=10))
```
